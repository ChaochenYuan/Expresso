```python
import awkward as ak
from coffea.nanoevents import NanoEventsFactory, BaseSchema, NanoAODSchema
import coffea.hist as hist
import matplotlib.pyplot as plt
```

```python
def getdataset(loc,label):
    
    events = NanoEventsFactory.from_root(loc,schemaclass=NanoAODSchema,treepath='Events',
                                         metadata={"dataset": label}).events()
    
    
    return events
```

```python
events=getdataset("/scratch/G2/tree_1.root","G2 sample")
```

```python
el = hist.Hist("Events",hist.Bin("genpartpdgID", "genpartpdgID", 50,0,50))
el.fill(genpartpdgID=ak.flatten(events.GenPart.pdgId))
ax = hist.plot1d(el)
```

```python
isphoton=(events.GenPart.pdgId==22)
events["genphotons"]=(events.GenPart[isphoton])
events["properphotons"]=events["genphotons"][abs(events.GenPart[events["genphotons"].genPartIdxMother].pdgId)==15]
events["properphotons"]=events["properphotons"][properphotons.status==1]
events["G2event"]=ak.num(events["properphotons"])>0
```

```python
h = hist.Hist("Events",hist.Bin("genpartpdgID", "genpartpdgID", 50,0,50))
h.fill(genpartpdgID=ak.flatten(events["properphotons"].pdgId))
ax = hist.plot1d(h)
```
